```python
import numpy as np
from matplotlib import pyplot as plt
from math import sqrt
```

```python
w = np.linspace(0,60,500)
```

```python
U = 1
w0 = 47.14*1e3
beta = 12*1e3
def f(x):
    x*=1000
    return (U*w0*w0)/(sqrt((w0*w0-x*x)**2+4*(beta**2)*x*x))
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(w, np.vectorize(f)(w), color='red',zorder=1)
plt.xticks(np.arange(0,61,2.5),fontsize=6)
for i in np.arange(0,61,2.5):
    plt.plot([i,i],[1,2.2],linewidth=0.5,color='grey',zorder=-10)
for i in np.arange(1,2.2,0.025):
    plt.plot([0,60],[i,i],linewidth=0.5,color='grey',zorder=-10)
plt.yticks(np.arange(1,2.2,0.025),fontsize=5) 
plt.scatter(43.98,f(43.98),color='blue',zorder=2)
plt.plot([43.98,43.98],[1,2.05],linestyle='dashed',color='orange',zorder=1)
plt.plot([0,43.98],[f(43.98),f(43.98)],linestyle='dashed',color='orange',zorder=1)
fig = plt.gcf()
plt.draw()
plt.show()
fig.savefig('res_phys2.png',dpi=1000)
```
